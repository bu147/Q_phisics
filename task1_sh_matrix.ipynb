```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Матрица гамильтониана

## свободное движение, $v = 0$

```python
size = 20
h = np.eye(size) * 2 - np.eye(size, k=1) - np.eye(size, k=-1)
print(h)
```

```python
#Собственные значения и вектора соответственно
eig, eve = np.linalg.eig(h)
print(eig)
```

```python
fig, ax = plt.subplots(figsize=(20, 10))
ax.scatter([i for i in range(size)], sorted(eig), s=10)

ax.set_xlabel('n',fontsize=25)
ax.set_ylabel('$%s$'%'E_n',fontsize=25)
plt.show()
```

### Пример существования локального уровня

```python
size = 100

# номер локального уровня
lvl = 50
h = np.eye(size) * 2 - np.eye(size, k=1) - np.eye(size, k=-1)
h[lvl, lvl] -= 1
eig, eve = np.linalg.eig(h)

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(20, 20))
ax1.scatter([i for i in range(size)], sorted(eig), s=8)
ax1.set_xlabel('n',fontsize=25)
ax1.set_ylabel('$%s$'%'E_n',fontsize=25)
ax1.axhline(y = 0, color='black', lw=.5)

ax2.plot([i for i in range(size)], eve.T[0]**2)
ax2.set_xlabel('n',fontsize=25)
ax2.set_ylabel('$%s$'%'\\Psi_n^*\\Psi_n',fontsize=25)
plt.show()
```

### Пример существования двух локальных уровней

```python
size = 1000

lvl1= 500
lvl2 = 450
h = np.eye(size) * 2 - np.eye(size, k=1) - np.eye(size, k=-1)
h[lvl1, lvl1] -= 5
h[lvl2, lvl2] -= 10
eig, eve = np.linalg.eig(h)

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(20, 20))
ax1.scatter([i for i in range(size)], sorted(eig), s=8, color='red')
ax1.set_xlabel('n',fontsize=25)
ax1.set_ylabel('$%s$'%'E_n',fontsize=25)
ax1.axhline(y = 0, color='black', lw=.5)
ax2.plot([i for i in range(size)], eve.T[0]**2)
ax2.plot([i for i in range(size)], eve.T[1]**2)
ax2.set_xlabel('n',fontsize=25)
ax2.set_ylabel('$%s$'%'\\Psi_n^*\\Psi_n',fontsize=25)
plt.show()
```

### Пример формирования уровней в потенциальной яме

```python
size // 2 - width
```

```python
size = 1000

# ширина ямы
width = 20
h = np.eye(size) * 2 - np.eye(size, k=1) - np.eye(size, k=-1)
left = size // 2 - width
right = size // 2 + width
for i in range(left, right + 1):
    h[i, i] -= 4
eig, eve = np.linalg.eig(h)

fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(20, 30))
ax1.plot([i for i in range(size)], np.diag(h) - 2)
ax1.set_xlabel('n',fontsize=25)
ax1.set_ylabel('$%s$'%'V_n',fontsize=25)

ax2.scatter([i for i in range(size)], sorted(eig), s=8, color='red')
ax2.set_xlabel('n',fontsize=25)
ax2.set_ylabel('$%s$'%'E_n',fontsize=25)
ax2.axhline(y = 0, color='black', lw=.5)

for i in range(width * 2 + 1):
    ax3.plot([i for i in range(size)], eve.T[i]**2)

ax3.set_xlabel('n',fontsize=25)
ax3.set_ylabel('$%s$'%'\\Psi_n^*\\Psi_n',fontsize=25)
plt.show()
```
